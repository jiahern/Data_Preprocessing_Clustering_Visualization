```python
#import ....
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import csv
import random
import sklearn
from scipy.spatial.distance import pdist, squareform
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA as sklearnPCA
from numpy import vstack,array
from numpy.random import rand
from scipy.cluster.vq import kmeans,vq
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist

#VAT algorithm provided for your use
def VAT(R):
    """

    VAT algorithm adapted from matlab version:
    http://www.ece.mtu.edu/~thavens/code/VAT.m

    Args:
        R (n*n double): Dissimilarity data input
        R (n*D double): vector input (R is converted to sq. Euclidean distance)
    Returns:
        RV (n*n double): VAT-reordered dissimilarity data
        C (n int): Connection indexes of MST in [0,n)
        I (n int): Reordered indexes of R, the input data in [0,n)
    """
        
    R = np.array(R)
    N, M = R.shape
    if N != M:
        R = squareform(pdist(R))
        
    J = list(range(0, N))
    
    y = np.max(R, axis=0)
    i = np.argmax(R, axis=0)
    j = np.argmax(y)
    y = np.max(y)


    I = i[j]
    del J[I]

    y = np.min(R[I,J], axis=0)
    j = np.argmin(R[I,J], axis=0)
    
    I = [I, J[j]]
    J = [e for e in J if e != J[j]]
    
    C = [1,1]
    for r in range(2, N-1):   
        y = np.min(R[I,:][:,J], axis=0)
        i = np.argmin(R[I,:][:,J], axis=0)
        j = np.argmin(y)        
        y = np.min(y)      
        I.extend([J[j]])
        J = [e for e in J if e != J[j]]
        C.extend([i[j]])
    
    y = np.min(R[I,:][:,J], axis=0)
    i = np.argmin(R[I,:][:,J], axis=0)
    
    I.extend(J)
    C.extend(i)
    
    RI = list(range(N))
    for idx, val in enumerate(I):
        RI[val] = idx

    RV = R[I,:][:,I]
    
    return RV.tolist(), C, I

```

```python
### Question 2.1

print("***")

import json

# Load the file
with open('roads.json',"r") as f:
    #convert json to python object
    file= json.load(f) 
    # select the directory to the part we need
    objects = file["meta"]["view"]["columns"]
    columns = list(objects)
    

# Select out the items which is in fieldName
column_names = [col["fieldName"] 
                for col in columns]

# Select out the columns that appropriate for us.
good_columns = column_names[8:]

# Select out all the appropriate data from json meta file
data = []
with open('roads.json',"r") as f:
    file = json.load(f)
    objects = file['data']
    for row in objects:
        selected_row = []
        for item in good_columns:
            selected_row.append(row[column_names.index(item)])
            data.append(selected_row)

street = pd.DataFrame(data, columns = good_columns)

# Drop off the irrelevant columns
street.drop(street.columns[[0,2,3,4,5,7,8,9]],axis = 1, inplace=True)
# Drop the duplicate values in specific column
street.drop_duplicates(subset ="seg_id",keep = 'first',inplace = True)

street["seg_id"] = pd.to_numeric(street["seg_id"])

# Merge both of the dataframe
traffic = traffic.merge(street,left_on = "road_segment", right_on = "seg_id")
print("The first three rows of traffic DataFrame with the attribute StrType are:")
traffic.head(3)

```

```python
### Question 2.2
print("***")

print("Q2.2")

# Create new column by doing reduction between two columns
traffic['max_speed_over_limit'] = traffic.apply(lambda row: row['maximum_speed'] - row['speed_limit'], axis=1)
print("The first three rows of traffic DataFrame with the new max_speed_over_limit attribute are:")
traffic.head(3)

```

```python
### Question 2.3

print("***")
print("Q2.3")
print("Three Arterial roads with the highest maximum max_speed_over_limit:")

# Slice out the row which strtype is Arterial
arterials = traffic[traffic.str_type == "Arterial"]

# Find out the highest max_speed_over_limit
max_grouped_arterials = arterials.groupby("road_name").max()

# Sort it descending
max_grouped_arterials.sort_values("max_speed_over_limit", ascending = False,inplace = True)

# Print out the value
for i in max_grouped_arterials.head(3).index:
    print(i,":",max_grouped_arterials.loc[i,"max_speed_over_limit"])

#   With this information, VicRoads is able to figure out the roads where vehicles exceeded speed limit the most,
# and install more notice board about the speed limit and speed trap camera along the roads which has more speeding vechiles.   
# This may increase the income of Vicroads issueing more traffic ticket. Lastly, it creates a better Victoria, 
# a place which is safe for Victoria's residents. 


```

```python
### Question 3.2 a

# Read the csv file and randomly input 1000 entries 
special_traffic = pd.read_csv("special_traffic.csv").sample(n=1000)

# Slice out the required columns
special_traffic_2 = special_traffic.iloc[:,1:19]

# Normalize special traffic_2
normalized_special_traffic_2 = sklearn.preprocessing.normalize(special_traffic_2)

# The first two PC
sklearn_pca = sklearnPCA(n_components=2)   

special_traffic_2_sklearn = sklearn_pca.fit_transform(normalized_special_traffic_2)

# Normalize two component of special_traffic_2
normalized_special_traffic_2_sklearn = sklearn.preprocessing.normalize(special_traffic_2_sklearn)

# Plot out the scatter plot
palette = palette = ["blue","green"]

colors = special_traffic.StrType.replace(to_replace=special_traffic.StrType.unique(),value=palette).tolist()

plt.scatter(special_traffic_2_sklearn[:,0],special_traffic_2_sklearn[:,1],s=60,c=colors)
plt.xlabel("1st Principal Component",fontsize=16)
plt.ylabel("2nd Principal Component",fontsize=16)
plt.show()
```

```python
### Question 3.2 b
#   From my perspective, the scatter plots show how much one variable is affected by another. The x_axis be the average speed, 
# y_axis be the vehicle_class_1. These two columns have the largest variance. Majority of vertical cluster's average speed 
# is lower while majority of horizontal cluster's average speed is higher. In addition, majority of vertical cluster's 
# vehicles' volume is higher while majority of horizontal cluster's vehicle's volume is lower.Hence, it can be analysed that 
# the vertical cluster is Council Minor while horizontal cluster is Arterial.
```

```python
### Question 3.3
print("Q3.3")
print("a")


# k means determine k
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(normalized_special_traffic_2)
    kmeanModel.fit(normalized_special_traffic_2)
    distortions.append(sum(np.min(cdist(normalized_special_traffic_2, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / normalized_special_traffic_2.shape[0])

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of Clusters')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()


```

```python
###Questoin 3.3 a
#   Yes, it is the expected value from Q3.2 since in 3.2 there are two cluster, and the optimal k is 2 as well, 
# so it is appropriate. This is because after 2, the distance between the next value is getting closer.
```

```python
###Question 3.3 b

#Apply data with K-means clustering and choose k=3 which is given.
centroids,_ = kmeans(normalized_special_traffic_2,3)
idx,_ = vq(normalized_special_traffic_2,centroids)

#Get the array of the clusters
size_cluster = np.bincount(idx)

#Plot the Bar Plot
Clusters = ("cluster 1","cluster 2","cluster 3")
y_pos = np.arange(len(Clusters))
plt.bar(y_pos, size_cluster, align='center', alpha=0.5)
plt.xticks(y_pos, Clusters)
plt.ylabel('quantity')
plt.title('Size of Each Cluster Bar Plot')

plt.show()
```

```python
### Question 3.3 c
#   From my perspective, I would recommend Scatter plot with K-means clustering. This is because there are three cluster. 
# With scatter plot, it allowed large data to be analysed more appropriately and viewed more detailly. 
```

```python
### Question 3.3 d

# Set the k-means value = 3
kmeans = KMeans(n_clusters=3)
kmeans.fit(normalized_special_traffic_2)

## Plot out the data with k-means value = 2 
plt.scatter(normalized_special_traffic_2[:,0],normalized_special_traffic_2[:,1], label='True Position',c=kmeans.labels_, cmap='rainbow')

# The center point of the two clusters in black dot
plt.scatter(kmeans.cluster_centers_[:,0] ,kmeans.cluster_centers_[:,1], color='black')
```

```python
### Question 3.3 d

#   It can be clearly seen that the plot has three distinct colours which indicate the similarity of each feature. 
# Besides that, with the black dot it indicates the center point of the clusters. Each of the cluster's point is spreaded around
# the center black point. However, there are few points are not within the clusters. Hence, Scatter plot is an appropriate way 
# of visualising. 
```
